#evaluation

```
from gensim.models import KeyedVectors
model =  KeyedVectors.load_word2vec_format("entity_vector/entity_vector.model.bin", binary=True)
#model =  KeyedVectors.load_word2vec_format("jawiki.entity_vectors.300d.txt")
```

```
results = model.wv.most_similar(positive=['仙台'])
for result in results:
    print(result)
```

```
results = model.wv.most_similar(positive=['東北大学','関東'], negative=['東北'],)
for result in results:
    print(result)
```

```
results = model.wv.most_similar(positive=['ビール', '日本'], negative=['ドイツ'],)
for result in results:
    print(result)
```
