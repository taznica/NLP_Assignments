```
from transformers import pipeline
text_generation_ja = pipeline(
    "text-generation",
    model="geekfeed/gpt2_ja",
    tokenizer="geekfeed/gpt2_ja",
)
text_generation_en = pipeline(
    "text-generation",
    model="gpt2",
    tokenizer="gpt2",
)
```

```
text_generation_ja("東北大学は")
```

```
text_generation_ja("情報科学研究科とは")
```

```
text_generation_en("Oh my God, that is horrendous!")
```
